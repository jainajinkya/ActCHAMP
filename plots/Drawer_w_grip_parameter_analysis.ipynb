```python
#!/usr/bin/env python3
import numpy as np
import transformations as tf
```

```python
def find_axis_in_camera_frame(pt, ref):
    quat = ref[3:]
    t = tf.quaternion_matrix(quat)
    pt = np.append(pt, 1.)
    new_pt = t.dot(pt)[:3]
    return new_pt
```

```python
def find_axis_in_global_frame(pt):
    quat = np.array([ 0., -0.707, 0., 0.707])
    t = tf.quaternion_matrix(quat)
    pt = np.append(pt, 1.)
    new_pt = t.dot(pt)[:3]
    return new_pt
```
