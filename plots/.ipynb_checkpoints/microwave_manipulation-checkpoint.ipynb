```python
import numpy as np
import matplotlib.pyplot as plt
import pickle as pkl

np.set_printoptions(formatter={'float_kind':'{:f}'.format})
```

```python
## Data for microwave Opening 30 deg
y2_bel = np.array([0., 0.32, 0.488, 0.523, 0.5235, 0.5235, 0.5235,
                   0.5234, 0.5235, 0.5235])
y2_err = np.array([0.500000, 0.166667, 0.125000, 0.119048, 0.118182, 
                   0.118056, 0.118037, 0.118034, 0.118034, 0.118034])

y2_gt = np.array([0.087266, 0.3, 0.462, 0.488, 0.488, 0.488, 
                  0.488, 0.488, 0.489, 0.489])

print("y2_plan = ", y2_bel)
print("y2_gt = ", y2_gt)
print("y2_err = ", y2_err)
```

```python
## Data for microwave closing 30 deg
y3_bel = np.array([0.649, 0.349, 0.149, 0.12, 0.08, 0.07, 
                  0.05, 0.05, 0.05, 0.05])
y3_err = np.array([0.5, 0.4969, 0.3956, 0.2896, 0.2043, 0.118182, 
                   0.118056, 0.118037, 0.118034, 0.118034])
y3_gt = np.array([0.489, 0.189, 0.15, 0.12, 0.11 , 0.11,
                  0.11, 0.11, 0.11, 0.11])

print("y3_plan = ", y3_bel)
print("y3_err = ", y3_err)
print("y3_gt = ", y3_gt)
```

```python
## Data for 50 deg
y4_bel = np.array([0.087, 0.387, 0.688, 0.835, 0.872, 0.872,
                   0.872, 0.872, 0.872, 0.872])
y4_err = np.array([0.5, 0.4969, 0.3956, 0.2896, 0.2043, 0.118182, 
                   0.118056, 0.118037, 0.118034, 0.118034])
y4_gt = np.array([0.11, 0.375, 0.601, 0.755, 0.79, 0.802,
                  0.802, 0.802, 0.802, 0.802])

print("y4_plan = ", y4_bel)
print("y4_err = ", y4_err)
print("y4_gt = ", y4_gt)
```

```python
fig = plt.figure(figsize=(40,25))
x = np.arange(1, 11, 1)


## Belief Trajectory
y = y4_bel
y_err = y4_err
y_gt = y4_gt

(_, caps, _) = plt.errorbar(x, y, yerr=y_err, label='Belief Trajectory', 
             fmt='o-', markersize=30, capsize=40, linewidth=7.0)


## Actual Trajectory
plt.plot(x, y_gt, label="Actual Trajectory", linewidth=7.0)

## Plot start and end points
plt.scatter([x[0]], [y[0]], c="k", marker="s", s=[3000])
plt.scatter([x[-1]], [y[-1]], c="g", marker="D", s=[3000])

plt.legend(loc='lower right', prop={'size': 50})
ax = plt.gca()
ax.tick_params(axis='both', which='major', labelsize=50)

for cap in caps:
    cap.set_markeredgewidth(5)
```

```python
err2 = abs(y2_gt[-1] - y2_bel[-1])
err3 = abs(y3_gt[-1] - y3_bel[-1])
err4 = abs(y4_gt[-1] - y4_bel[-1])
err5 = abs(y5_gt[-1] - y5_bel[-1])
err6 = abs(y6_gt[-1] - y6_bel[-1])

errs = np.array([err2, err3, err4, err5, err6])

print("Error2 : ", err2)
print("Error3 : ", err3)
print("Error4 : ", err4)
print("Error5 : ", err5)
print("Error6 : ", err6)
print("Mean_err : ", np.mean(errs))
print("std_err : ", np.std(errs))

## Delta Q
delta_2 = abs((y2_bel[-1] - y2_bel[0])/(y2_gt[-1] - y2_gt[0]))
delta_3 = abs((y3_bel[-1] - y3_bel[0])/(y3_gt[-1] - y3_gt[0]))
delta_4 = abs((y4_bel[-1] - y4_bel[0])/(y4_gt[-1] - y4_gt[0]))
delta_5 = abs((y5_bel[-1] - y5_bel[0])/(y5_gt[-1] - y5_gt[0]))
delta_6 = abs((y6_bel[-1] - y6_bel[0])/(y6_gt[-1] - y6_gt[0]))

print("\n")
print("Delta 2 : ", delta_2)
print("Delta 3 : ", delta_3)
print("Delta 4 : ", delta_4)
print("Delta 5 : ", delta_5)
print("Delta 6 : ", delta_6)

deltas = np.array([delta_2, delta_3, delta_4, delta_5, delta_6])
print("Mean_delts : ", np.mean(deltas))
print("std_deltas : ", np.std(deltas))
```
