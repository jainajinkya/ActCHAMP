```python
with open('../pomdp_hd/traj_data/drawer_run6_delta_neg_15.pkl', 'rb') as f:
    data = pkl.load(f, encoding='latin1')

## Data for Drawer
y6_bel = np.array(data['mu_plan'][0]).astype(float)
y6_err = np.array(data['s_plan'][0])
y6_err[0] = 100.0
y6_err /= 100.
y6_err += 0.5

y6_gt = np.array([6.1, 16.8, 22.1, 22.1, 22.1, 22.1, 22.1, 22.1, 22.1, 22.1])
y6_gt = 22.1 - y6_gt

print("y6_plan = ", y6_bel)
print("y6_err = ", y6_err)
print("y6_gt = ", y6_gt)
```

```python
fig = plt.figure(figsize=(40,25))
x = np.arange(1, 11, 1)


## Belief Trajectory
# y = y6_bel
# y_err = y6_err
# y_gt = y6_gt

y = y2_bel
y_err = y2_err
y_gt = y2_gt


(_, caps, _) = plt.errorbar(x, y, yerr=y_err, label='Belief Trajectory', 
             fmt='o--', markersize=50, capsize=60, linewidth=15.0)


## Actual Trajectory
plt.plot(x, y_gt, label="Actual Trajectory", linewidth=15.0)

## Plot start and end points
plt.scatter([x[0]], [y[0]], c="g", marker="s", s=[3000])
plt.scatter([x[-1]], [y[-1]], c="g", marker="D", s=[3000])

plt.scatter([x[0]], [y_gt[0]], c="k", marker="s", s=[3000])
plt.scatter([x[-1]], [y_gt[-1]], c="k", marker="D", s=[3000])

## Settings
plt.xlabel('t', fontsize=100)
plt.ylabel('$q$', fontsize=100)
plt.legend(loc='lower right', prop={'size': 100})
plt.tight_layout()

ax = plt.gca()
ax.tick_params(axis='both', which='major', labelsize=80)
ax.grid(True, linewidth=4)
for axis in ['top','bottom','left','right']:
  ax.spines[axis].set_linewidth(5.)


for cap in caps:
    cap.set_markeredgewidth(5)
```

```python
err2 = abs(y2_gt[-1] - y2_bel[-1])
err3 = abs(y3_gt[-1] - y3_bel[-1])
err4 = abs(y4_gt[-1] - y4_bel[-1])
err5 = abs(y5_gt[-1] - y5_bel[-1])
err6 = abs(y6_gt[-1] - y6_bel[-1])

errs = np.array([err2, err3, err4, err5, err6])

print("Error2 : ", err2)
print("Error3 : ", err3)
print("Error4 : ", err4)
print("Error5 : ", err5)
print("Error6 : ", err6)
print("Mean_err : ", np.mean(errs))
print("std_err : ", np.std(errs))

## Delta Q
delta_2 = abs(err2/(y2_gt[-1] - y2_gt[0]))
delta_3 = abs(err3/(y3_gt[-1] - y3_gt[0]))
delta_4 = abs(err4/(y4_gt[-1] - y4_gt[0]))
delta_5 = abs(err5/(y5_gt[-1] - y5_gt[0]))
delta_6 = abs(err6/(y6_gt[-1] - y6_gt[0]))

print("\n")
print("Delta 2 : ", delta_2)
print("Delta 3 : ", delta_3)
print("Delta 4 : ", delta_4)
print("Delta 5 : ", delta_5)
print("Delta 6 : ", delta_6)

deltas = np.array([delta_2, delta_3, delta_4, delta_5, delta_6])
print("Mean_delts : ", np.mean(deltas))
print("std_deltas : ", np.std(deltas))
```
